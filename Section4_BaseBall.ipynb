## Q) KBO 최고의 타자는?

### 선수별 기록 집계

```python
# 피벗 테이블_선수별 기록 집계
data_player = data.pivot_table(index=['팀', '이름', '생일'],
                               values=['타수', '안타', '홈런', '루타', '타점', '볼넷', '사구', '희비'],
                               aggfunc='sum') # aggfunc default opt: avg

data_player
```

```python
# 지표 비교 위해 타수가 너무 적은 선수를 제외하기 위한 타수 데이터 분포 확인

data_player['타수'].hist()
```

```python
# 타수>50
cond = data_player['타수']>50

data_player[cond]
```

```python
data_player = data_player[cond].reset_index()
data_player.head()
```

**<지표>**
- 타율: 타격에 성공해서 진루하는 비율 **= 타격 성공 횟수 / 타격 기회 수**
- 출루율: 살아서 진루하는 비율 **= 진루 성공 횟수 / 진루 기회 수**
- 장타율: 타율에 진루한 베이스 가중치 추가 **= 진루한 베이스 수 / 타격 기회 수 = 루타 수 / 타수**
- OPS **= 출루율 / 장타율**

```python
# 지표 계산 함수
def cal_hit(df):
    df['타율'] = df['안타'] / df['타수']
    df['출루율'] = (df['안타'] + df['볼넷'] + df['사구']) / (df['타수'] + df['사구'] + df['희비'])
    df['장타율'] = df['루타'] / df['타수']
    df['OPS'] = df['출루율'] + df['장타율']
    return df
```

```python
player_stat = cal_hit(data_player)
player_stat.head()
```

```python
# KBO 타자를 성적순으로 정렬

# 기준: 출루율-장타율-OPS-타율
player_stat = player_stat.sort_values(by=['출루율', '장타율', 'OPS', '타율'], ascending=False) #내림차순
player_stat = player_stat.reset_index(drop=True)
player_stat
```

### 상대팀별 기록 정리하기

```python
raw['상대'].unique()
```

```python
# 상대 컬럼에서 @가 있으면 원정경기, 없으면 홈경기
# '상대' 컬럼에서 홈/원정 경기, 상대팀을 분리하여 '홈어웨이', '상대팀' 컬럼으로 지정
opp_list = []
home_away_list = []

for opp in raw['상대']:
    if '@' in opp:
        home_away = '원정'
        opp = opp.replace('@', '')

    else:
        home_away = '홈'

    opp_list.append(opp)
    home_away_list.append(home_away)

raw['홈어웨이'] = home_away_list
raw['상대팀'] = opp_list
```

```python
raw.head(10)
```

```python
# 상대팀별 실적 정리
data = raw.pivot_table(index=['팀', '이름', '생일', '상대팀'],
                       values=['타수', '안타', '홈런', '루타', '타점', '볼넷', '사구', '희비'],
                       aggfunc='sum')
data.head()
```

```python
cond = data['타수']>0
data = data[cond]

data.head()
```

```python
data = data.reset_index()
data.head()
```

```python
# 타자/상대팀별 실적 계산
player_stat_opp = cal_hit(data)
player_stat_opp.head()
```
